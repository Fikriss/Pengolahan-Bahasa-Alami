```python
import pandas as pd

# Load the dataset dengan mencoba berbagai opsi separator
file_path = '/content/df_yesdok_stemmed.csv'  # Ganti dengan path file yang sesuai

# Coba membaca tanpa parameter sep
try:
    data_flexible = pd.read_csv(file_path)
except pd.errors.ParserError:
    print("ParserError: Mencoba dengan separator ';'")
    data_flexible = pd.read_csv(file_path, sep=';', error_bad_lines=False)

# Tampilkan data yang berhasil dimuat
print(data_flexible.head())
```

```python
# 1. Find reviews that mention 'dokter' or 'konsultasi'
dokter_reviews = data_flexible[data_flexible['content'].str.contains(r'(dokter|konsultasi)', case=False, na=False)]
print("Reviews mentioning 'dokter' or 'konsultasi':")
dokter_reviews.head()
```
